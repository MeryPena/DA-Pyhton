**The Interquartile Range** (IQR)

Is defined as the range between the 75th percentile (Q3)
and the 25th percentile (Q1) because these percentiles capture the *middle 50%*
of the data.

**Why IQR Method?**

The mean and standard deviation are sensitive to extreme values, meaning they can be
heavily distorted by outliers.
The IQR, on the other hand, focuses on *the central part of the data*, making it more
robust against outliers and skewed distributions.

```python
#Import Numpy library
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# Example list of given values
dataset = [15,22,22,25,30,35,40,40,40,50,55,60,200,220]


# Calculate mean
mean_value = np.mean(dataset)

# Calculate standard deviation
std_population = np.std(dataset, ddof=0)  # Population standard deviation
std_sample = np.std(dataset, ddof=1)  # Sample standard deviation

print(f"Population Standard Deviation: {std_population:.2f}")
print(f"Sample Standard Deviation: {std_sample:.2f}")


# 1st Step: Calculation for percentile 25% & 75%
# The 25th percentile (Q1) is the value below which 25% of the data falls.
# The 75th percentile (Q3) is the value below which 75% of the data falls.

Q1 = np.percentile(dataset, 25)
Q3 = np.percentile(dataset, 75)
print("percentile 25 Q1:", Q1, ",", "percentile 75 Q3:",Q3)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Dataset
dataset = [15, 22, 22, 25, 30, 35, 40, 40, 40, 50, 55, 60, 200, 220]

# Calculate statistics
mean_value = np.mean(dataset)
std_dev = np.std(dataset, ddof=1)  # Sample standard deviation

# Plot distribution
plt.figure(figsize=(10, 5))
sns.histplot(dataset, bins=10, kde=True, color="skyblue", edgecolor="black")

# Plot mean line
plt.axvline(mean_value, color="red", linestyle="--", label=f"Mean: {mean_value:.2f}")

# Plot standard deviation lines
plt.axvline(mean_value + std_dev, color="green", linestyle="--", label=f"+1 SD: {mean_value + std_dev:.2f}")
plt.axvline(mean_value - std_dev, color="green", linestyle="--", label=f"-1 SD: {mean_value - std_dev:.2f}")

plt.axvline(mean_value + 2*std_dev, color="purple", linestyle="--", label=f"+2 SD: {mean_value + 2*std_dev:.2f}")
plt.axvline(mean_value - 2*std_dev, color="purple", linestyle="--", label=f"-2 SD: {mean_value - 2*std_dev:.2f}")

plt.title("Data Distribution with Mean and Standard Deviation")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Dataset
dataset = [15, 22, 22, 25, 30, 35, 40, 40, 40, 50, 55, 60, 200, 220]

# Compute Q1, Q3, and IQR
Q1 = np.percentile(dataset, 25)  # 25th percentile
Q3 = np.percentile(dataset, 75)  # 75th percentile
IQR = Q3 - Q1

# Define outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Detect outliers
outliers = [x for x in dataset if x < lower_bound or x > upper_bound]

# Create a box plot
plt.figure(figsize=(10, 4))
sns.boxplot(x=dataset, color="lightblue", flierprops={"marker": "o", "markerfacecolor": "red", "markersize": 8})
plt.title("Box Plot for Outlier Detection (IQR Method)")
plt.xlabel("Values")
plt.show()

# Scatter plot highlighting outliers
plt.figure(figsize=(10, 4))
sns.scatterplot(x=range(len(dataset)), y=dataset, color="blue", label="Data Points")
sns.scatterplot(x=[dataset.index(x) for x in outliers], y=outliers, color="red", label="Outliers", s=100)
plt.axhline(y=lower_bound, color="gray", linestyle="--", label="Lower Bound")
plt.axhline(y=upper_bound, color="gray", linestyle="--", label="Upper Bound")
plt.title("Outliers in the Dataset (IQR Method)")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend()
plt.show()
```

```python
import numpy as np

# Example list of values
data = [15, 22, 22, 25, 30, 35, 40, 40, 40, 50, 55, 60, 200, 220]

# Step 1: Calculate Q1 and Q3
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)

# Step 2: Compute IQR
IQR = Q3 - Q1

# Step 3: Determine Outlier Boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Step 4: Remove Outliers
filtered_data = [x for x in data if lower_bound <= x <= upper_bound]

print("Original Data:", data)
print("Filtered Data (Without Outliers):", filtered_data)
```
